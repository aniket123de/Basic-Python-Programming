## **1. Bitwise AND -**

```python
X = 5
Y = 12
Z = X & Y
print(Z)
```

## **2. Bitwise OR -**

```python
X = 5
Y = 13
Z = X | Y
print(Z)
```

## **3. Bitwise XOR -**

```python
X = 13
Y = 21
Z = X ^ Y
print(Z)
```

## **4. Bitwise Left-shift -**

```python
x = 16
y = x << 2
print(y)
```

## **5. Bitwise Right-shift -**

```python
x = 16
y = x >> 2
print(y)
```

## **6. Negation -**

```python
x = 16
y = ~x
print(y)
```

## **Methods in String -**

### **1. capitalize() -**

```python
x = "aniket de"
print(x.capitalize())
```

*It will capitalize only the first character of the string input.*

### **2. center() -**

```python
x= "aniket"
print(x.center(20))
```

*The string "aniket" is centered, with 7 spaces on the left and 7 spaces on the right*

### **3. count() -**

```python
x = "I am a good boy who is also good in study"
y = "good"
ans = x.count(y)
print(ans)
```

### **4. endswith() -**

```python
x = "I am a good boy who is also good in study"
y = "good"
ans = x.endswith(y)
print(ans)
ans = x.endswith("boy")
print(ans)
```

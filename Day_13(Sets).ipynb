## **Properties of Set**



*   Set is Unordered
*   Set is unindexed
*   Set contains distinct element





**Set is Unordered.**

```python
x = {100,20,30,4}
print(x)
```

*Output is not as same as input.*

**Set contains distinct element**

```python
x = {100,20,30,4,20,30}
print(x)
```

*If we repeat same element , at the time of same element no duplicate value will be there.Output contains distinct element.*

**Set is unindexed**

```python
x = {100,20,30,4}
print(x[2])
```

*Solution : Convert set into list.*

```python
x = {100,20,30,4}
x = list(x)
print(x[2]) # Order isn't same
```

## **Set Methods**

### **1. update()**

```python
x = {10,20,30}
y = {'Kolkata','Mumbai','Delhi'}
x.update(y)
print(x)
```

*Here , x and y both are sets.*

*   *x.update(y) => x=x+y*
*   *i.e. x will be updated by x+y*



### **2. add()**

```python
x = {10,20,30}
y = ('Kolkata','Mumbai','Delhi')
x.add(y)
print(x)
```

```python
x = {10,20,30}
y = ['Kolkata','Mumbai','Delhi']
x.add(y)
print(x)
```

```python
x = {10,20,30}
y = {'Kolkata','Mumbai','Delhi'}
x.add(y)
print(x)
```

*Only tuple can be added by add()*

### **3. remove()**

```python
x = {10,20,30,40}
print(x)
x.remove(20)
print(x)
```

*i.e. Set is mutable/changeable.*

```python
x = {10,20,30,40}
print(x)
x.remove(50)
print(x)
```

If the value isn't present , then it will throw - KeyError

### **4. discard()**

```python
x = {10,20,30,40}
print(x)
x.discard(20)
print(x)
```

```python
x = {10,20,30,40}
print(x)
x.discard(70)
print(x)
```

*remove() vs discard()*

*   *remove => If the value isn't present , it'll throw keyerror*
*   *discard => If the value isn't present , it will not throw keyerror*



### **5. pop()**

```python
x = {10,20,50,40}
print(x)
x.pop()
print(x)
```

*It removes a random element from the set.*

```python
x = {10,20,50,40}
x.pop()
x.pop()
x.pop()
x.pop()
x.pop()
```

*If we operate pop() more than the number of elements , it'll throw KeyError.*

### **8. issubset()**

```python
x = {10,20,30,40,50}
y = {10,20,30}
print(x.issubset(y))
print(y.issubset(x))
print(x.issubset(x))
print(y.issubset(y))
```

### **9. issuperset()**

```python
x = {10,20,30,40,50}
y = {10,20,30}
print(x.issuperset(y))
print(y.issuperset(x))
print(x.issuperset(x))
print(y.issuperset(y))
```

### **10. union()**

```python
x = {10,20,30,40,50}
y = {10,20,70,80}
print(x.union(y))
print(y.union(x))
print(x)
print(y)
```

```python
x = {10,20,30,40,50}
y = {10,20,70,80}
print(x|y) # Through operator
```

### **11. intersection()**

```python
x = {10,20,30,40,50}
y = {10,20,70,80}
print(x.intersection(y))
print(y.intersection(x))
print(x)
print(y)
```

```python
x = {10,20,30,40,50}
y = {10,20,70,80}
print(x&y) # Through operator
```

### **12. intersection_update()**

```python
x = {10,20,30,40,50}
y = {10,20,70,80}
print(x)
print(y)
x.intersection_update(y) # x = x&y
print(x)
print(y)
```

### **13. difference()**

```python
x = {10,20,30,40,50}
y = {10,20,70,80}
print(x)
print(y)
print(x.difference(y))
print(y.difference(x))
print(x)
print(y)
```

```python
x = {10,20,30,40,50}
y = {10,20,70,80}
print(x-y) # Through operator
```

### **14. difference_update()**

```python
x = {10,20,30,40,50}
y = {10,20,70,80}
print(x)
print(y)
x.difference_update(y) # x = x-y
print(x)
print(y)
```

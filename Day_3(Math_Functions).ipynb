# **LCM of 2 Numbers :**

Multiplication of a and b = a*b = gcd(a,b) * lcm(a,b)

```python
a = int(input())
b = int(input())
print(a,b)
x = math.gcd(a,b)
lcm = (a*b)/x
print(lcm)
```

Check a number is a square number or not ?
Example - 17 is not a sq number , 25 is a sq number.

```python
# First method
import math
x = int(input())
if (math.sqrt(x))**2==x:
  print('Square Number')
else:
  print('Not Square Number')
```

```python
# Another method
import math
a=int(input())
if(math.sqrt(a)==math.isqrt(a)):
  print("Square")
else:
  print("Not Square")
```
